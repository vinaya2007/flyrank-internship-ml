## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# Research Question

## Search Question
How do you choose what content pages should be reviewed first based on signals from search and engagement?

## Unit of Analysis
Each content page is represented one time.

## Output
A list of pages ranked by priority for content refresh and reasons given for the ranking.

## Decision
To help SEO or content specialists to understand what pages need to be reviewed first.

## Action
To review the most important pages first and establish if any content changes, updates, optimization, or just monitoring is required.

## Cost of a Wrong Recommendation
Choosing a page that is not really important would cause the team to waste valuable time in the review process. Not choosing an important page would lead to observation opportunities being lost. Thus, recommendations should help in making an independent decision rather than being decision-makers.

## Why Machine Learning Can Help
The rule-based approach is simplistic as it takes into account one or two indicators at most. On the contrary, ML can discover correlations between various observable indicators (for instance, impressions, CTR, engagement, content age, and average position) to perform better at ranking while still allowing people to make the final decision.
```
